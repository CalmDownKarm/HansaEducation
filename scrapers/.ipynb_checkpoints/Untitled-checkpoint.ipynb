```python
import json
with open('../Output/Dirty Shiksha/ShikshaUSAMSFinal.json') as data_file:
    data = json.load(data_file)
```

```python
all_keys = set().union(*(d.keys() for d in data))
```

```python
def lowercasekeys(course):
    return dict((k.lower(), v) for k,v in course.items())
```

```python
lowercased = list(map(lowercasekeys,data))
```

```python
all_keys = set().union(*(d.keys() for d in lowercased))
```

```python
all_keys
```

```python
books_string_inr = [x for x in all_keys if (('books' in x) and ('inr' in x))]
books_string_usd = [x for x in all_keys if (('books' in x) and ('usd' in x))]
books_string_inr +=['course materials inr ','study costs and books inr ','textbooks inr ','study equipment (including books) inr ']
books_string_usd +=['course materials usd ','study costs and books usd ','textbooks usd ','study equipment (including books) usd ']
living_string_inr = [x for x in all_keys if (('living' in x) and ('inr' in x))]
living_string_usd = [x for x in all_keys if (('living' in x) and ('usd' in x))]
misc_string_inr = [x for x in all_keys if (('miscellaneous' in x) and ('inr' in x))]
personal_expense_inr = [x for x in all_keys if (('personal' in x) and ('inr' in x))]
misc_string_usd = [x for x in all_keys if (('miscellaneous' in x) and ('usd' in x))]
personal_expense_usd = [x for x in all_keys if (('personal' in x) and ('usd' in x))]
academic_inr = [x for x in all_keys if (('academic' in x) and ('inr') in x)]
academic_usd = [x for x in all_keys if (('academic' in x) and ('usd') in x)]
activity_inr = [x for x in all_keys if (('activity' in x) and ('inr') in x)]
activity_usd = [x for x in all_keys if (('activity' in x) and ('usd') in x)]
campus_inr = [x for x in all_keys if (('campus' in x) and ('inr') in x)]
campus_usd = [x for x in all_keys if (('campus' in x) and ('usd') in x)]
Computer_inr = [x for x in all_keys if (('Computer' in x) and ('inr') in x)]
Computer_usd = [x for x in all_keys if (('Computer' in x) and ('usd') in x)]
enrollment_inr = [x for x in all_keys if (('enrollment' in x) and ('inr') in x)]
enrollment_usd = [x for x in all_keys if (('enrollment' in x) and ('usd') in x)]
health_inr = [x for x in all_keys if (('health' in x) and ('inr') in x)]
health_usd = [x for x in all_keys if (('health' in x) and ('usd') in x)]


```

```python
def clean_up_keys(school):
    bitch = {}
    for k,v in school.items():
        if str(k) in books_string_inr:
            bitch["books cost inr"] = v
        elif str(k) in books_string_usd:
            bitch["books cost usd"] = v
        elif str(k) in living_string_inr:
            bitch["living costs inr"] = v
        elif str(k) in living_string_usd:
            bitch["living costs usd"] = v
        elif str(k) in misc_string_inr:
            bitch["misc costs inr"] = v
        elif str(k) in living_string_usd:
            bitch["living costs usd"] = v
        elif str(k) in personal_expense_inr:
            bitch["personal expenses inr"] = v
        elif str(k) in personal_expense_usd:
            bitch["personal expenses usd"] = v
        else:
            bitch[str(k)] = v
    return bitch
```

```python
cleaned = list(map(clean_up_keys,lowercased))
```

```python
all_keys_2 = set().union(*(d.keys() for d in cleaned))
```
