```python
from bs4 import BeautifulSoup as bs
import pandas as pd
import csv
```

```python
def extract_urls(path):
    try:
        with open(path) as html_file:
            return [(link.get_text(),link.get('href')) for link in bs(html_file.read(),"html.parser").find_all('a')]
    except IOError:
        return []
```

```python
result = extract_urls('FT details.html')
```

```python
import csv
with open('NamesAndLinks.csv','w') as out:
	csv_out = csv.writer(out)
	csv_out.writerows(result)
```

```python
df = pd.DataFrame(result,columns = ['Name','Link'])
```

```python
df
```

```python
import urllib.request
def extract_table(url):
	with urllib.request.urlopen(url) as page:
		s = page.read()
	htmlsoup = bs(s,"html.parser")
	tables = htmlsoup.find_all("table",{"class":"entitytable halfwidth"})
	vals = []
	for table in tables:
		for row in table.findAll('tr'):
			col = row.findAll('td')
			if col:
				heading = col[0].get_text()
				val = col[1].find('span').get_text()
				vals.append(val)    
	return vals
```

```python
df['dataextract'] = df['Link'].apply(extract_table)
df
```

```python
new_df = df.copy()
```

```python
new_col_list = ['Rank in 2017',
'Rank in 2016',
'Rank in 2015',
'3 year average rank',
'Country',
'Audit year',
'Salary today (US$)',
'Weighted salary (US$)',
'Salary percentage increase',
'Value for money rank',
'Career progress rank',
'Aims achieved (%)',
'Careers service rank',
'Employed at three months (%)',
'Alumni recommend rank',
'Female faculty (%)',
'Female students (%)',
'Female board (%)',
'International faculty (%)',
'International students (%)',
'International board (%)',
'International mobility rank',
'International course experience rank',
'Languages',
'Faculty with doctorates (%)',
'PhD graduates No. (%)',
'FT research rank']
```

```python
for n,col in enumerate(new_col_list):
    new_df[col] = new_df['dataextract'].apply(lambda dataextract: dataextract[n])
```
